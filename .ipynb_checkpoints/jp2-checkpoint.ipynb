```python
%%time
def ipfn(ip,ip2):
    lst = []
    fs,sc,thr,four = ip.split('.')
    fs2,sc2,thr2,four2 = ip2.split('.')
    fsz = (int(fs)*(256**3))+(int(sc)*(256 **2))+(int(thr)*256)+int(four)
    fsz2 =(int(fs2)*(256**3))+(int(sc2)*(256 **2))+(int(thr2)*256)+int(four2)
    sum2=(random.randint((1),(2**100))% (fsz2 - fsz))+fsz
    d = int(sum2 % 256)
    c = int(((sum2%(256**2))-d)/256)
    b = int((sum2%(256 **3)-d-(c*256))/(256**2))
    a = int(sum2 / (256**3))
    lst.append(str(a))
    lst.append(str(b))
    lst.append(str(c))
    lst.append(str(d))
    return '.'.join(lst)
dic =[]
with open('./CN-20161122.txt') as fd:
    for lines in fd.readlines():
        fs = lines.split()[0]
        end = lines.split()[1]
        dic.append([fs,end])
ipset = set()
while len(ipset) < 50000:
    r=random.randint(0,len(dic)-1)
    ipset.add(ipfn(dic[r][0],dic[r][1]))
```

```python
 def Npoland(exp):
        rule = {'+': 1, '-': 1, '*': 2, '/': 2, '(': 1, ')': -1}
        ret = []
        for j in re.split(r'(\W+?)', exp):
            if j is '':
                continue
            if j.isdigit():
                ret.append(j)
            else:
                if len(stack) == 0 or rule[j] >= rule[stack[-1]]:
                    stack.append(j)
                elif j is ')':
                    while True:
                        if stack[-1] is '(':
                            stack.pop()
                            break
                        else:
                            ret.append(stack.pop())
                else:
                    while len(stack) > 0:
                        ret.append(stack.pop())
                    stack.append(j)
        else:
            while len(stack) > 0:
                ret.append(stack.pop())
        return ret
```

```python
stack = []
for i in Npoland('9+(3-1)*3+10/2'):
    if i.isdigit():
        stack.append(int(i))
    else:
        if i is '+':
            stack.append(stack.pop(-2) + stack.pop())
        elif i is '-':
            stack.append(stack.pop(-2) - stack.pop())
        elif i is '*':
            stack.append(stack.pop(-2) * stack.pop())
        elif i is '/':
            stack.append(stack.pop(-2) / stack.pop())
print(stack[-1])
```
